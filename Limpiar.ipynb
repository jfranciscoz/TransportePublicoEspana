```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
df = pd.read_csv('Transporte_Publico_Espana.csv',na_values='not_available')  # Read the csv with "not_available" as a NaN type as well
```

```python
df.info()
```

## Duplicados:

```python
df.duplicated().sum()
```

```python
df_duplicated = df[df['ID_Viaje'].duplicated(keep='first')]
duplicated_IDs = df_duplicated['ID_Viaje'].tolist()
print(duplicated_IDs)
```

```python
df = df.drop_duplicates(subset='ID_Viaje', keep='first')
```

```python
df.duplicated().sum()
```

## Tipo de datos:

```python
df['Fecha'] = pd.to_datetime(df['Fecha'])
df['Ruta'] = df['Ruta'].astype(str)
df['Tipo_Transporte'] = df['Tipo_Transporte'].astype(str)
df['Region'] = df['Region'].astype(str)
df['Dia_Semana'] = df['Dia_Semana'].astype(str)
df.dtypes
```

## Missing values:

```python
pasajeros = df['Numero_Pasajeros'].isna().sum()
duracion = df['Duracion_Viaje_Minutos'].isna().sum()
print(f"Numero_Pasajeros tiene {pasajeros} ({pasajeros/100000*100}%) missing values.")
print(f"Duracion_Viaje_Minutos tiene {duracion} ({duracion/100000*100}%) missing values.")
```

```python
df_pasajeros_missing = df[df['Numero_Pasajeros'].isna()==1]
# df_pasajeros_missing['Dia_Semana'].value_counts()  # Missing data homogeneously distributed across days of de week
# df_pasajeros_missing['Region'].value_counts()  # Missing data is more or less homogeneously distributed across regions
```

Considering only a 1.66% of data will be lost. We will drop the rows with missing values.

```python
df.dropna(inplace=True)
df.isnull().sum()
```
